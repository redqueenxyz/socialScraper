```python
results = nfl.interest_over_time()
#results = results.append(nflcont.interest_over_time())
```

```python
# Iterate and append

suggestions = []

for key in nfl.related_queries():
    #print(key)
    
    result = nfl.related_queries()[key]['top']['query'].tolist()
    if result:
        suggestions += result
    else: 
        pass
    
print(suggestions)
```

```python
new_trends = TrendReq(google_username, google_password)

# Final Set

for i in range(0,len(suggestions)//5):
    new_words = suggestions[i:i+4]
    print(new_words)
    
    new_trends.build_payload(new_words, geo="US", timeframe="2016-01-01 2016-09-01") #2016-09-01 2017-11-01
    results = pandas.concat([results, new_trends.interest_over_time()], axis=1)
```

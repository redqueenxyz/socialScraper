## Analysis

Analyze results and store them in dataframes

Create dataframes to recieve results

```r
names(coreresults)
```

```r
trendresults = data.frame(coreresults[[1]]$trend[1:2])
risingresults = data.frame(index = 1:10) #nrow(productset[[1]]$rising[[1]])
searchresults = data.frame(index = 1:15) #nrow(productset[[1]]$searches[[1]])
```

```r
for (page in 1:length(coreresults)) {
    # Summarize Trend Results
    trendset  = coreresults[[page]]$trend[3]
    trendresults = cbind(trendresults, trendset)
    # debug: print(head(set))
}
```

```r
print(head(trendresults))
```

```r
for (page in 1:length(coreresults)) {
    if (length(coreresults[[page]]$rising) > 0) {
        risingset  = coreresults[[page]]$rising[[1]]
        risingresults = cbind(risingresults, risingset[1:10,])
    } else {
    next
    }    
    #risingresults = cbind(risingresults, risingset[1:10])
}
```

```r
print(risingresults)
```

```r
for (page in 1:length(coreresults)) {
    if (length(coreresults[[page]]$searches) > 0) {
        searchset  = coreresults[[page]]$searches[[1]]
        searchresults = cbind(searchresults, searchset[1:15,])
    } else {
    next
    }    
}
```

```r
print(searchresults)
```
